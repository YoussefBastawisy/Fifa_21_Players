# Split joined into Day , Month , Year

```python
df["Joined"].fillna("NO DATE" , inplace = True)
```

```python
df["Joined"].isnull().sum()
df["Joined"].values.astype(str)
```

```python
df[["Month" , "ehm"]]= df["Joined"].str.split(" ", 1 , expand = True)
```

```python
df[["day" , "year"]]= df["ehm"].str.split(",", 1 , expand = True)
```

```python
df = df.drop(columns=['ehm','Joined'])
```

# Change datatype of Value and Wage into integers

```python
df[["Value" , "Wage" ]].astype(int,errors='ignore').fillna(0 ,inplace=True)
```

```python
df.fillna( 0 , inplace = True)
```

```python
df= df.drop(columns="Unnamed: 0")
```

# Delete (?) from Weak Foot , Skill Moves and International Reputation

```python
df["Weak Foot"]=df["Weak Foot"].str.replace('[?]', '')
df["Skill Moves"]=df["Skill Moves"].str.replace('[?]', '')
df["International Reputation"]=df["International Reputation"].str.replace('[?]', '')
```

# Delete unnecessary columns

```python
df = df.drop(columns= ["Likes" , "Dislikes"])
```
